```python
score = step_data(numpy.ndarray)
average = step_data(int)

@step
def generate_scores(score):
    score = numpy.random.randint(100, size=30)

@step
def calculate_curve():
    generate_scores(score)
    average = numpy.average(score)
    
calculate_curve()
```
